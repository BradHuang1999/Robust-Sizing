```python
import os
import numpy as np
from subprocess import run, PIPE
import matplotlib.pyplot as plt
```

```python
os.chdir('../../')
os.getcwd()
```

```python
def run_robust_sizing(binary, pv_cost, battery_cost, pv_max, cell_max,
                      metric, epsilon, conf, n_days, load_file, pv_file, seed=''):
    
    cmd = f'{binary} {pv_cost} {battery_cost} {pv_max} {cell_max} {metric} {epsilon} {conf} {n_days} {load_file} {pv_file} {seed}'
    p = run(cmd, stdout=PIPE, stderr=PIPE, shell=True)
    
    if p.returncode == 0:
        return tuple(p.stdout.decode().split())
    else:
        raise Exception(f'bad return: {p.returncode}')
```

```python
inputs = {
    'binary': './bin/sim',
    'pv_cost': 2000,
    'battery_cost': 500,
    'metric': 1,
    'conf': 0.95,
    'n_days': 100,
    'load_file': 'example_inputs/load.txt',
    'pv_file': 'example_inputs/pv.txt',
    'pv_max': 70,
    'cell_max': 225,
    'epsilon': 0.01,
}

run_robust_sizing(**inputs, seed=10)
```

```python
positive_results = []
negative_results = []

seed = 0
while seed < 100000:
    try:
        result_tup = run_robust_sizing(**inputs, seed=seed)
        positive_results.append((seed, result_tup))
    except Exception as ex:
        negative_results.append((seed, ex))
    finally:
        seed += 1
```
