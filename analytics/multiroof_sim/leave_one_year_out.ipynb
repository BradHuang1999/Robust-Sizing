```python
for epsilon, df_split in df_loo.groupby('epsilon'):
    print(f'Îµ={epsilon:.2f}')
#     epsilon = 1 * epsilon
#     print(f'new_epsilon={epsilon:.2f}')
    print()
    
    fig, (ax_vloss, ax_cost, ax_duration) = plt.subplots(nrows=1, ncols=3, figsize=(20, 4))
    fig.suptitle(f'Îµ={epsilon:.2f}')
    
    ax_vloss.set_title('histogram of validation loss')
    ax_vloss.set_xlabel('validation loss')

    ax_cost.set_title('histogram of cost')
    ax_cost.set_xlabel('cost')

    ax_duration.set_title('histogram of computation time')
    ax_duration.set_xlabel('seconds')
    
    lgds = []
    arr_vloss = []
    arr_cost = []
    arr_duration = []
    
    for conf, df_conf in df_split.groupby('confidence'):
        print(f'conf={conf:.2f}')
        lgds.append(f'conf={conf:.2f}')
        
        den_conf_feasibles = df_conf[df_conf['feasible'] == 1]
        den_feasible = len(den_conf_feasibles) / len(df_conf)
        print(f'observed_density(feasible): {den_feasible:.3f}')
        
        den_loss = len(den_conf_feasibles[den_conf_feasibles['validation_loss'] > epsilon]) / len(df_conf)
        print(f'observed_density(loss > {epsilon:.2f}): {den_loss:.3f}')
        
        print(f'mean(cost): {np.mean(den_conf_feasibles["cost"]):.3f}')
        print()
        
        arr_vloss.append(den_conf_feasibles['validation_loss'].reset_index(drop=True))
        arr_cost.append(den_conf_feasibles['cost'].reset_index(drop=True))
        arr_duration.append(den_conf_feasibles['duration'].reset_index(drop=True))
        
    ax_vloss.hist(arr_vloss, alpha=0.5, density=True)
    ax_vloss.axvline(x=epsilon, color='black')
    ax_cost.hist(arr_cost, alpha=0.5, histtype='bar')
    ax_duration.hist(arr_duration, alpha=0.5, histtype='bar')

    plt.legend(lgds)
    plt.show()
```

```python
font = {'size'   : 16}
matplotlib.rc('font', **font)

fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(9, 7), dpi=200)

texts = [
#     (0, 0, ''),
    (0.11, 7, '''Î³=0.85: 4.1%
  sizings with loss > 0.1
Î³=0.95: 1%
  sizings with loss > 0.1'''),
    (0.08, 1.7, '''Î³=0.85: 2.6%
  sizings with loss > 0.5
Î³=0.95: 2%
  sizings with loss > 0.5'''),
]

for i, (epsilon, df_split) in enumerate(df_loo.groupby('epsilon')):
    if epsilon == 0.05:
        continue
        
    i -= 1    
    
    print(f'Îµ={epsilon:.2f}')
    print()
    
    axes[i].set_title(f'Îµ={epsilon:.2f}')
    axes[i].set_xlabel('validation loss')
    axes[i].set_ylabel('count')
    
    lgds = []
    arr_vloss = []
    
    for conf, df_conf in df_split.groupby('confidence'):
        if conf in [0.75, 0.97]:
            continue
            
        print(f'Î³={conf:.2f}')
        lgds.append(f'Î³={conf:.2f}')
        
        den_conf_feasibles = df_conf[df_conf['feasible'] == 1]
        den_feasible = len(den_conf_feasibles) / len(df_conf)
        print(f'observed_density(feasible): {den_feasible:.3f}')
        
        den_loss = len(den_conf_feasibles[den_conf_feasibles['validation_loss'] > epsilon]) / len(df_conf)
        print(f'observed_density(loss > {epsilon:.2f}): {den_loss:.3f}')
        
        print(f'mean(cost): {np.mean(den_conf_feasibles["cost"]):.3f}')
        print()
        
        print(f'mean(total):\n{den_conf_feasibles[["pv1", "pv2", "battery", "cost"]].mean()}')
        print()
        
        print(f'quantile(loss, {conf}): {np.quantile(den_conf_feasibles["validation_loss"], conf)}')
        print()
        
#         print(f'df(pv1=0):\n{den_conf_feasibles[den_conf_feasibles["pv1"] == 0]}')
#         print()
                
#         print(f'df(pv2=0):\n{den_conf_feasibles[den_conf_feasibles["pv2"] == 0]}')
#         print()
        
        arr_vloss.append(den_conf_feasibles['validation_loss'].reset_index(drop=True))
        
    axes[i].hist(arr_vloss, bins=12, rwidth=0.618, alpha=0.8, density=True)
    axes[i].legend(lgds)
    axes[i].axvline(x=epsilon, lw=2, color='red')
    axes[i].text(*texts[i]).set_fontsize(13)

plt.subplots_adjust(hspace=0.5)
plt.show()
```

```python
df_by = df_by.reset_index()
```

```python
df_by = df_by[(df_by['epsilon'].isin([0.1, 0.5])) & (df_by['confidence'].isin([0.85, 0.95]))]
```

```python
font = {'size'   : 15}
matplotlib.rc('font', **font)

prms = [
    ('pv1', 'pv1', '(# of panels)'),
    ('pv2', 'pv2', '(# of panels)'),
    ('battery', 'Battery', '(kWh)'),
    ('thousands_cost', 'Cost', '(thousand $)')
]

fig, axes = plt.subplots(nrows=len(prms), ncols=1, figsize=(7, 1.8*len(prms)), sharex=True, dpi=200)
tot_width = 0.25

for i, (col, lbl, unit) in enumerate(prms):
    axes[i].set_xticks([0.5, 0.1])
    axes[i].set_xlim(0.7, -0.1)

    gbys = df_by.groupby(['confidence'])
    for j, (conf, df_conf) in enumerate(gbys):
        if conf in [0.75, 0.97]:
            continue

        ep_x = df_conf['epsilon'] - tot_width / len(gbys) / 2 + tot_width / len(gbys) * (len(gbys) - j - 1)
        conf_y = df_conf[col]
        rects = axes[i].bar(ep_x, conf_y, tot_width / len(gbys), label=f'Î³={conf}')
        for rect in rects:
            height = rect.get_height()
            axes[i].annotate(f'{height:.0f}',
                        xy=(rect.get_x() + rect.get_width() / 2, height),
                        xytext=(0, 3),  # 3 points vertical offset
                        textcoords="offset points",
                        ha='center', va='bottom')
            
    axes[i].set_ylim(ymin=0, ymax=1.25 * axes[i].get_ylim()[1])
    axes[i].text(0.97, axes[i].get_ylim()[1] / 3, lbl + "\n" + unit)
        
axes[-1].set_xlabel('Îµ')
axes[0].legend(bbox_to_anchor=(0.5, 1.5), loc='upper center', ncol=2)

fig.subplots_adjust(hspace=0.2)
fig.show()
```
